```python
# Create 18 district dictionary
district = {
  'District_cn': ['中西區', '九龍城區', '元朗區', '北區', '南區', '大埔區', '屯門區', '東區', '沙田區', '油尖旺區', '深水埗區', '灣仔區', '荃灣區', '葵青區', '西貢區', '觀塘區', '離島區', '黃大仙區'],
  'District_en': ['Central and Western District', 'Kowloon City District', 'Yuen Long District', 'North District', 'Southern District', 'Tai Po District', 'Tuen Mun District', 'Eastern District', 'Sha Tin District', 'Yau Tsim Mong District', 'Sham Shui Po District', 'Wan Chai District', 'Tsuen Wan District', 'Kwai Tsing District', 'Sai Kung District', 'Kwun Tong District', 'Islands District', 'Wong Tai Sin District']
}

#Create Dataframe
df_district = pd.DataFrame.from_dict(district)

df_restaurant = df_grouped_restaurants
df_mall = df_grouped_malls
df_garden = df_grouped_gardens
df_clinic = df_pet_clinic_grouped
df_housing = df_grouped_housing
df_cat_growth = df_cat_growth
df_dog_growth = df_dog_growth
df_series = df_series
```

```python
#Merge all facilities table
df_all_facilities = df_district.merge(df_restaurant, how='left', on='District_en').fillna(0)
df_all_facilities = df_all_facilities.merge(df_mall, how='left', on='District_en').fillna(0)
df_all_facilities = df_all_facilities.merge(df_garden, how='left', on='District_cn').fillna(0)
df_all_facilities = df_all_facilities.merge(df_clinic, how='left', on='District_cn').fillna(0)
df_all_facilities = df_all_facilities.merge(df_housing, how='left', on='District_en').fillna(0)

# Output the facilities table as csv
df_all_facilities.to_csv('facilities_table.csv', index=False, encoding='utf-8-sig')
df_all_facilities
```

```python
# Merge facilities table with dogs number on 2023 (by district)
df_dog = df_dog_growth[['distric','2023']].round()
df_dog_2023 = df_dog.rename(columns={'distric':'District_en', '2023': 'Dog_2023'})
df_dog_2023['District_en'] = df_dog_2023['District_en'].str.rstrip()
df_dogs_facilities = df_all_facilities.merge(df_dog_2023, how='left', on='District_en')

#Calculate the ratio of each facilities for dogs by district
df_dogs_facilities['Restaurant'] = df_dogs_facilities.apply(lambda x:x['Dog_2023']/x['Restaurant'] if x['Restaurant'] != 0 else 0, axis=1)
df_dogs_facilities['Mall'] = df_dogs_facilities.apply(lambda x:x['Dog_2023']/x['Mall'] if x['Mall'] != 0 else 0, axis=1)
df_dogs_facilities['Pet Garden'] = df_dogs_facilities.apply(lambda x:x['Dog_2023']/x['Pet Garden'] if x['Pet Garden'] != 0 else 0, axis=1)
df_dogs_facilities['Garden Area(M2)'] = df_dogs_facilities.apply(lambda x:x['Dog_2023']/x['Garden Area(M2)'] if x['Garden Area(M2)'] != 0 else 0, axis=1)
df_dogs_facilities['Clinic'] = df_dogs_facilities.apply(lambda x:x['Dog_2023']/x['Clinic'] if x['Clinic'] != 0 else 0, axis=1)
df_dogs_facilities['Housing Estate'] = df_dogs_facilities.apply(lambda x:x['Dog_2023']/x['Housing Estate'] if x['Housing Estate'] != 0 else 0, axis=1)

#Export as csv
df_dogs_facilities = df_dogs_facilities.round()
df_dogs_facilities.to_csv('dog_facilities_ratio.csv', index=False, encoding='utf-8-sig')
df_dogs_facilities
```

```python
# Merge facilities table with cats number on 2023 (by district)
df_cat = df_cat_growth[['distric','2023']].round()
df_cat_2023 = df_cat.rename(columns={'distric':'District_en', '2023': 'Cat_2023'})
df_cat_2023['District_en'] = df_cat_2023['District_en'].str.rstrip()
df_cat_facilities = df_all_facilities.merge(df_cat_2023, how='left', on='District_en')

# Calculate the ratio of each facilities for dogs by district
df_cat_facilities['Restaurant'] = df_cat_facilities.apply(lambda x:x['Cat_2023']/x['Restaurant'] if x['Restaurant'] != 0 else 0, axis=1)
df_cat_facilities['Mall'] = df_cat_facilities.apply(lambda x:x['Cat_2023']/x['Mall'] if x['Mall'] != 0 else 0, axis=1)
df_cat_facilities['Pet Garden'] = df_cat_facilities.apply(lambda x:x['Cat_2023']/x['Pet Garden'] if x['Pet Garden'] != 0 else 0, axis=1)
df_cat_facilities['Garden Area(M2)'] = df_cat_facilities.apply(lambda x:x['Cat_2023']/x['Garden Area(M2)'] if x['Garden Area(M2)'] != 0 else 0, axis=1)
df_cat_facilities['Clinic'] = df_cat_facilities.apply(lambda x:x['Cat_2023']/x['Clinic'] if x['Clinic'] != 0 else 0, axis=1)
df_cat_facilities['Housing Estate'] = df_cat_facilities.apply(lambda x:x['Cat_2023']/x['Housing Estate'] if x['Housing Estate'] != 0 else 0, axis=1)

#Export as csv
df_cat_facilities = df_cat_facilities.round()
df_cat_facilities.to_csv('cat_facilities_raio.csv', index=False, encoding='utf-8-sig')
df_cat_facilities
```
